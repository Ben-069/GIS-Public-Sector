PARAGUAY'S YUCA PRODUCTION 2021

```python
yucapy=pd.read_excel('https://docs.google.com/uc?id=1k-osFo8Jrx84omMmw6jXDt57yXEO-s4I&export=download', header=4)
```

```python
yucapy.head(25)
```

```python
pip install unidecode
```

```python
import pandas as pd
from unidecode import unidecode
```

```python
# Supongamos que la columna que quieres normalizar se llama 'nombre'
yucapy['NAME1_'] = yucapy['Desagregaciones'].apply(lambda x: unidecode(x) if isinstance(x, str) else x)

# Guardar el dataset con la columna normalizada en un nuevo archivo Excel
#yucapy.to_excel("tu_archivo_normalizado.xlsx", index=False)

#I got this code from AI
```

```python
# Eliminar los asteriscos al final de las palabras
yucapy['NAME1_'] = yucapy['NAME1_'].str.replace(r'\s*\*', '', regex=True)
```

```python
#This is the new dataset with the normalized characters
yucapy2021=yucapy[['NAME1_', 'Unnamed: 11']]
```

```python
yucapy2021.head(25)
```

```python
yucapy2021['NAME1_'] = yucapy2021['NAME1_'].str.replace('Pdte.', 'Presidente', regex=False)
```

```python
yucapy2021.dtypes
```

```python
#merge part 1 (se combinan los datos del shapefile y el data y todos tienen que decir both (deseable))
mandioca2021= pd.merge(yucapy2021, agripyC0, on='NAME1_',how='outer',indicator=True) #(make sure to have njC shapefile from previous sec)
mandioca2021[['NAME1_','Unnamed: 11','_merge']] #all good, NEW JERSEY expected not to merge
```

```python
#merge part 2
mandioca_final = pd.merge(yucapy2021, agripyC0, on='NAME1_',how='inner',indicator=False) #Este codigo es para combinar solo aquellos que dicen both
```

```python
type(mandioca_final)
```

```python
gdmandioca=gpd.GeoDataFrame(mandioca_final)
```

```python
fig, ax = plt.subplots(1, figsize=(6,10))
gdmandioca.plot(ax=ax, column='Unnamed: 11', legend=True, cmap='YlGn',
                scheme='natural_breaks', k=5, edgecolor='black', linewidth=1,
                legend_kwds={"fmt": "{:,.0f}", 'loc': 'center left', 'bbox_to_anchor': (1, 0.5), 'markerscale': 1.4},
                missing_kwds={'color': 'grey', 'edgecolor': 'black'})

leg1 = ax.get_legend()
leg1.set_title('''Legend''')  # Cambié el texto a "Legend"
ax.set_title("Paraguay Yuca Production 2021 (ton)")
ax.set_xticks([])  # Elimina las etiquetas en el eje x
ax.set_yticks([])  # Elimina las etiquetas en el eje y
ax.set_xlabel('''MAG: MINISTERIO DE AGRICULTURA - PARAGUAY''')

# plt.savefig('fig.pdf', bbox_inches='tight')
```
